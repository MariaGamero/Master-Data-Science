**Second exercise: top 10 arrival airports in the world in 2013 (using the bookings file)**

```python
#filenames
fbookings=path_to_bz2 + 'bookings.csv.bz2'

fbreader=pd.read_csv(fbookings,delimiter='^',compression='bz2', usecols=['arr_port','year','pax'], chunksize=10000)

data = pd.DataFrame()
for chunk in fbreader:
    chunk = chunk[chunk['year']==2013]
    chunk.keys().str.strip()
    chunk['arr_port'] = chunk['arr_port'].str.strip().str.upper()
    data = data.append(chunk)

datafinal = data.groupby('arr_port')['pax'].sum().sort_values(ascending = False).head(10)
datafinal
```

```python
dfsearches2 = pd.read_csv(fsearches, delimiter='^', usecols=['Destination','Date'], nrows=100000, compression = 'bz2')
dfsearches2.sample(20)
```

```python
#formateo
dfsearches2['Destination']=dfsearches2['Destination'].str.strip().str.upper()
dfsearches2['Date']=dfsearches2['Date'].str.strip()

dfsearches2.sample(5)
```

```python
#Opcion 1: Extraer el mes y crear la columna month para filtrar
dfsearches2['Month']=dfsearches2['Date'].str[0:7]
dfsearches2.sample(5)
```

```python
#Opción 2: cambiar formato a pd.datetime. Lo incluimos en la columna date2 y agrupamos por mes.
```

```python
#Opción 3: leer el archivo leer esa columna con DateTime.
```

### Filtro el fichero por las tres ciudades (eliminando el resto)

```python
dfsearches3=dfsearches2[dfsearches2['Destination'].isin(['MAD','BCN','AGP'])]
#dfsearches2[dfsearches2['Destination']=='AGP'|'BCN'|'MAD'].
dfsearches3.sample(5)
```

```python
dfsearches4=dfsearches3.groupby(['Destination','Month']).count()
dfsearches4
```

```python
dfsearches4.columns
```

```python
#IMPORTANTE ARREGLAMOS EL DT y lo convertimos a uno NORMAL
dfsearches4.columns=['searches'] #hemos cambiado el nombre a la columna DATE
prepared_df=dfsearches4.reset_index() #ponemos los indices todos al mismo nivel
prepared_df
```

## Plotting

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
fig=plt.figure(figsize=(10,6))
ax=plt.subplot(111)

prepared_df[prepared_df['Destination']=='MAD'].plot(x='Month',y='searches',ax=ax,label='Madrid')
prepared_df[prepared_df['Destination']=='BCN'].plot(x='Month',y='searches',ax=ax,c='green',label='Barcelona')
prepared_df[prepared_df['Destination']=='AGP'].plot(x='Month',y='searches',ax=ax,c='blue',label='Malaga')

plt.xlabel(r'month')
plt.ylabel(r'searches')
plt.legend(loc='best')
plt.grid()
plt.show()
```

## Ahora con el fichero grande

```python
fsearches=path_to_bz2 + 'searches.csv.bz2'

searches=pd.read_csv(fsearches,delimiter='^',compression='bz2', usecols=['Destination','Date'], chunksize=100000)

#data = pd.DataFrame()
dataframes=[]

for chunk in searches:
    chunk = chunk.dropna() #quitar los ceros
    searches_subset=chunk[chunk['Destination'].isin(['MAD','BCN','AGP'])]
    
    searches_subset['PlotDate']=searches_subset['Date'].str.slice(0,7)
    arrivals_per_destination=searches_subset.groupby(['PlotDate','Destination']).count()
    arrivals_per_destination.columns=['Searches'] #renombro la columna (porque solo hay una)
    
    prepared_df=arrivals_per_destination.reset_index()
    dataframes.append(prepared_df)
    
result = pd.concat(dataframes).groupby(['PlotDate','Destination']).sum().reset_index().pivot('PlotDate','Destination','Searches')

result
```
